```python
#IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('botak.csv')
```

## BAR CHART

```python
plt.figure(figsize=(10, 6))
edu_avg = df.groupby("pekerjaan")['botak_prob'].mean().sort_values()
sns.barplot(x=edu_avg.index, y=edu_avg.values, palette="coolwarm")
plt.xlabel("Jenis Pekerjaan")
plt.ylabel("Rata-rata Probabilitas")
plt.title("Rata-rata Probabilitas Berdasarkan Pekerjaan")
plt.xticks(rotation=45)
plt.show()
```

## LINE CHART

```python
plt.figure(figsize=(10,6))
age_avg = df.groupby("umur")['botak_prob'].mean()
plt.plot(age_avg.index, age_avg.values, marker='o', linestyle='-',color='b')
plt.xlabel("Usia")
plt.ylabel("Rata-rata probabilitas")
plt.title("Tren Probabilitas Berdasarkan Usia")
plt.grid()
plt.show()
```

## PIE CHART

```python
plt.figure(figsize=(8,8))
pendidikan_counts = df['pendidikan'].value_counts()
plt.pie(pendidikan_counts, labels=pendidikan_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribusi Riwayat Pendidikan")
plt.show
```
